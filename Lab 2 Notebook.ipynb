Step response input to left motor displayed via velocity versus time.

```python
from matplotlib import pyplot

def readcsv(filename):
    data = {'Time': [],'Position':[], 'Velocity': []} # Initialize data dictionary assuming 3 columns
    # Read the CSV file
    # The first line is the header
    
    with open(filename, 'r') as file:
        headers = file.readline().strip().split(',')
        
        for line in file:
            values = line.strip().split(',')
            if len(values) >= 3:
                try:
                    time = float(values[0])
                    velocity = float(values[2])  
                    
                    data['Time'].append(time)
                    data['Velocity'].append(velocity)
                    
                except ValueError:
                    pass
                    
        return data

def main():
    filename = 'Test1-Left.csv'     # Replace with appropriate filename
    # Read the CSV file
    data = readcsv(filename)
    
    time = [t - data['Time'][0] for t in data['Time']]  # Normalize time to start from 0
    # Scale velocity to match the original units
    velocity = [v * (256 / 60) for v in data['Velocity']]  
    
    pyplot.figure(figsize=(10, 6)) 
    
    pyplot.plot(time, velocity, label='Motor Speed', color='b')
    
    pyplot.title('Step Response - Left Motor Speed')
    pyplot.xlabel('Time (s)')
    pyplot.ylabel('Velocity (scaled)')
    pyplot.legend()
    
    pyplot.grid(True)  
    
    pyplot.show()
    
if __name__ == "__main__":
    main()
```
